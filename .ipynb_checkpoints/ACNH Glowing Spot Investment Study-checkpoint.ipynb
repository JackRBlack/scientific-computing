```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
def expectation(x, alpha):
    return (3*alpha - 1)*x + 30000 * (1-alpha)
```

```python
def expectation_2(x, beta):
    alpha = beta * (99000 - x) / 89000
    return (3*alpha - 1)*x + 30000 * (1-alpha)
```

```python
x1 = np.linspace(0, 10000, 10001)
x2 = np.linspace(10001, 99000, 89000)
```

```python
alpha = 0.3
E1 = 2 * x1
E2 = expectation(x2, alpha)
```

```python
def demo(alpha):
    E1 = 2 * x1
    E2 = expectation(x2, alpha)
    plt.figure(figsize=(5, 2), dpi=200)
    plt.plot(x1, E1, c='k')
    plt.plot(x2, E2, c='k')
    plt.xlabel('Buried Bells')
    plt.ylabel('Expectation')
    plt.show()
    return
```

```python
interact(demo, alpha=(0, 1, 0.1));
```

```python
def demo_2(beta):
    E1 = 2 * x1
    E2 = expectation_2(x2, beta)
    plt.figure(figsize=(5, 2), dpi=200)
    plt.plot(x1, E1, c='k')
    plt.plot(x2, E2, c='k')
    plt.xlabel('Buried Bells')
    plt.ylabel('Expectation')
    plt.show()
    return
```

```python
demo_2(1)
```
