# Lindhard function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g_1D(x):
    return 0.5*np.log(np.abs((1 + x)/(1 - x)))/x

def g_2D(x):
    g = []
    for t in x:
        if t < 1:
            g.append(1)
        else:
            g.append(1 - np.sqrt(1 - 1 / t**2))
    return np.array(g)
    
def g_3D(x):
    return 0.5 - 0.25 / x * (1 - x**2) * np.log(np.abs((1 - x)/(1 + x)))
```

```python
x = np.linspace(0.01, 2, 300)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, g_1D(x))
plt.plot(x, g_2D(x))
plt.plot(x, g_3D(x))
plt.show()
```
