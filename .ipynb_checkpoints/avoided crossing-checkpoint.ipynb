```python
k = np.linspace(-1, 1, 200)
```

```python
V = np.array([[0, 1],
              [1, 0]])
```

```python
k_point = -1
```

```python
H = np.array([[k_point, 0],
              [0, -k_point]])
H = H + V
eigval, eigvec = np.linalg.eig(H)
```

```python
eigval, eigvec
```

```python
E_h, E_l = [], []
a = 0.02
for k_point in k:
    H = np.array([[k_point, 0],
                  [0, -k_point]])
    H = H + a * V
    eigval, eigvec = np.linalg.eig(H)
    E_h.append(np.max(eigval))
    E_l.append(np.min(eigval))
```

```python
plt.figure(figsize=(4, 3), dpi=200)
plt.plot(k, E_h)
plt.plot(k, E_l)
plt.ylim(-1.2, 1.2)
plt.xlabel('k')
plt.ylabel('energy')
plt.show()
```

```python
def anti_crossing_demo(a):
    k = np.linspace(-1, 1, 201)
    V = np.array([[0, 1],
                  [1, 0]])
    E_h, E_l = [], []
    
    for k_point in k:
        H = np.array([[k_point, 0],
                      [0, -k_point]])
        H = H + a * V
        eigval, eigvec = np.linalg.eig(H)
        E_h.append(np.max(eigval))
        E_l.append(np.min(eigval))
    E_h = np.array(E_h)
    E_l = np.array(E_l)
    
    plt.figure(figsize=(4, 3), dpi=200)
    plt.plot(k, E_h)
    plt.plot(k, E_l)
    plt.ylim(-1.2, 1.2)
    plt.xlabel('k')
    plt.ylabel('energy')
    plt.show()
    return
```

```python
anti_crossing_demo(0.2)
```

```python
interact(anti_crossing_demo, a = (0, 0.5, 0.02))
```

```python
def anti_crossing_demo_2(a, b):
    k = np.linspace(-np.pi, np.pi, 201)
    V = np.array([[0, 1],
                  [1, 0]])
    E_h, E_l = [], []
    
    for k_point in k:
        H = np.array([[np.cos(k_point), 0],
                      [0, -b*np.cos(k_point)]])
        H = H + a * V
        eigval, eigvec = np.linalg.eig(H)
        E_h.append(np.max(eigval))
        E_l.append(np.min(eigval))
    E_h = np.array(E_h)
    E_l = np.array(E_l)
    
    plt.figure(figsize=(4, 3), dpi=200)
    plt.plot(k, E_h)
    plt.plot(k, E_l)
    plt.ylim(-1.2, 1.2)
    plt.xticks(np.linspace(-np.pi, np.pi, 3), ['$-\pi$', '0', '$\pi$'])
    plt.xlabel('k')
    plt.ylabel('energy')
    plt.show()
    return
```

```python
interact(anti_crossing_demo_2, a = (0, 0.5, 0.02), b = (0.5, 1.5, 0.1))
```

```python
def anti_crossing_demo_2_eigvec(a, b):
    k = np.linspace(-np.pi, np.pi, 201)
    V = np.array([[0, 1],
                  [1, 0]])
    E_h, E_l = [], []
    Eigvec = []
    
    for k_point in k:
        H = np.array([[np.cos(k_point), 0],
                      [0, -b*np.cos(k_point)]])
        H = H + a * V
        eigval, eigvec = np.linalg.eig(H)
        E_h.append(np.max(eigval))
        E_l.append(np.min(eigval))
        Eigvec.append(eigvec)
    E_h = np.array(E_h)
    E_l = np.array(E_l)
    Eigvec = np.array(Eigvec)
    return E_h, E_l, Eigvec
```

```python
E_h, E_l, Eigvec = anti_crossing_demo_2_eigvec(0.1, 0.5)
```
