```python
import numpy as np
import matplotlib.pyplot as plt
import random
#import time
```

```python
def gaussian(x, x0):
    '''
        t0: related to pulse width
        omega0: carrier frequency
        a: chirping term
    '''
    envelope = np.exp(- x**2 / x0**2)
    return envelope

def random_noise(x, a, b, n, w):
    centers = random.sample(range(a, b), n)
    noises = x - x
    for center in centers:
        width = random.uniform(w*0.5, w*1.5)
        noises = noises + np.exp(- (x - center)**2 / width**2)
    return noises
    
```

```python
def autocorr(x):
    result = np.correlate(x, x, mode='full')
    return result[result.size // 2:]
```

```python
x = np.linspace(-100, 100, 1000)
signal = gaussian(x, 40)
noisy_signal = random_noise(x, -100, 100, 12, 4) * gaussian(x, 40)
```

```python
plt.figure(figsize=(12,5))
plt.rcParams.update({'font.size': 15})
plt.plot(x, signal)
plt.plot(x, noisy_signal)
plt.xlabel("x (arb. unit)")
plt.ylabel("amplitude (arb. unit)")
plt.legend(["without noise", "with noise"])
plt.savefig("noisy_signal.eps", dpi = 300, format = 'eps')
plt.show()
```
