```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# two-level system Boltzmann distribution

def p_0(T):
    return 1 / (1 + np.exp(- 1 / T))

def p_1(T):
    return 1 - 1 / (1 + np.exp(- 1 / T))
```

```python
T = np.linspace(0.01, 3, 300)
```

```python
plt.figure(figsize=(10,5))
plt.plot(T, p_0(T))
plt.plot(T, p_1(T))
plt.xlabel('$T$ / $(\Delta E / k_B)$')
plt.ylabel('probability')
plt.legend(['ground state', 'excitation state'])
plt.show()
```

```python
# three-level system Boltzmann distribution
# suppose E_1 - E_0 = E_2 - E_1 = \Delta E

def p_0(T):
    return 1 / (1 + np.exp(- 1 / T) + np.exp(- 2 / T))

def p_1(T):
    return np.exp(- 1 / T) / (1 + np.exp(- 1 / T) + np.exp(- 2 / T))

def p_2(T):
    return np.exp(- 2 / T) / (1 + np.exp(- 1 / T) + np.exp(- 2 / T))
```

```python
T = np.linspace(0.01, 10, 300)

plt.figure(figsize=(10,5))
plt.plot(T, p_0(T))
plt.plot(T, p_1(T))
plt.plot(T, p_2(T))
plt.xlabel('$T$ / $(\Delta E / k_B)$')
plt.ylabel('probability')
plt.legend(['ground state', '1st excitation state', '2nd excitation state'])
plt.show()
```
