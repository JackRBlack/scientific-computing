$$ \frac{\nu'}{\nu} = \frac{1}{1 + \frac{\displaystyle h\nu}{\displaystyle m_0 c^2}(1 - \cos \theta)} $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import physical_constant as pc
```

```python
# calculate the energy for standing electrons and convert to eV

E_0 = pc.m_e * pc.c ** 2
E_0 = pc.J_2_eV(E_0)
```

```python
(nu_L, nu_U) = (0, 10**22)
nu_0 = np.linspace(nu_L, nu_U, 1000)
hnu_0 = pc.J_2_eV(pc.h * nu_0)
```

```python
theta = np.pi / 6
ratio = 1 / (1 + hnu_0 / E_0 * (1 - np.cos(theta)))
```

```python
# Plot the figure
plt.figure(figsize=(10,5))
plt.plot(hnu_0 / 10**6, ratio)
plt.xlabel("photon energy / MeV")
plt.ylabel("$\nu / \nu_0$")
plt.show()
```

```python
hnu_0 = 0.5 * 10**6 # eV
theta = np.linspace(0, 2 * np.pi, 200)
hnu = hnu_0 / (1 + hnu_0 / E_0 * (1 - np.cos(theta)))

# Plot the figure
plt.figure(figsize=(10,5))
plt.plot(theta, hnu / 10 ** 6)
plt.plot(theta, (hnu_0 - hnu) / 10 ** 6)
plt.xlabel("theta")
plt.ylabel("energy / MeV")
plt.show()
```

```python
hnu_0 = 0.5 * 10**6 # eV
theta = np.linspace(0, 2 * np.pi, 200)
ratio_1 = 1 / (1 + 0.5 * 10 **6 / E_0 * (1 - np.cos(theta)))
ratio_2 = 1 / (1 + 0.05 * 10 **6 / E_0 * (1 - np.cos(theta)))
ratio_3 = 1 / (1 + 0.005 * 10 **6 / E_0 * (1 - np.cos(theta)))

# Plot the figure
plt.figure(figsize=(10,5))
plt.plot(theta, ratio_1)
plt.plot(theta, ratio_2)
plt.plot(theta, ratio_3)
plt.legend(["E = 500 keV", "E = 50 keV", "E = 5 keV"])
plt.xlabel("theta")
plt.ylabel("ratio")
plt.show()
```
