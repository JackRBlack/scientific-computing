```python
import numpy as np
from matplotlib import pyplot as plt
```

## PDW d-wave form factor

$$
\Delta_{\mathbf{P}}(\mathbf{k}) = 2\Delta_{\mathbf{P}} \left[\cos(k_x - P_x/2) - \cos(k_y - P_y/2)\right]
$$

```python
Delta = 1
P_x, P_y = np.pi/3, np.pi/3
k_x = np.linspace(-np.pi, np.pi, 100)
k_y = np.linspace(-np.pi, np.pi, 100)
```

```python
K_X, K_Y = np.meshgrid(k_x, k_y, sparse=False, indexing='ij')
```

```python
OP = 2 * Delta * (np.cos(K_X - P_x/2) - np.cos(K_Y - P_y/2))
```

```python
plt.figure(figsize = (3, 3), dpi=200)
plt.pcolormesh(K_X, K_Y, OP) 
plt.show()
```
