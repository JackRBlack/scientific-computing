```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def XY_A(g, eta_A, eta_B, r):
    tc = np.cosh(2*r)
    ts = np.sinh(2*r)
    result = 1 - eta_A + eta_A * tc + g**2 * (1 - eta_B + eta_B * tc) + 2 * g * np.sqrt(eta_A * eta_B) * ts
    return result
```

```python
def XY_A_min(eta_A, eta_B, r):
    tc = np.cosh(2*r)
    result = 1 - eta_A * (tc - 1) * (2 * eta_B - 1) / (1 - eta_B + eta_B * tc)
    return result
```

```python
XY_A_min(0.2, 0.2, 2)
```

```python
eta = np.linspace(0, 1, 100)

plt.figure(figsize=(7,5))
plt.plot(eta, XY_A_min(eta, eta, 2), 'k-')
plt.plot(eta, XY_A_min(1, eta, 2), 'b--')
plt.plot(eta, XY_A_min(eta, 1, 2), 'r-.')

plt.xlim(0, 1)
plt.ylim(0, 1.6)

plt.legend(["$\eta_A = \eta_B = \eta$", "$\eta_A = 1, \eta_B = \eta$", "$\eta_A = \eta, \eta_B = 1$"])

plt.xlabel("$\eta$")
plt.ylabel("$\Delta_{inf} X^A \Delta_{inf} Y^A$")

#plt.savefig("EPR-efficiency.eps", dpi = 300, format = 'eps')

plt.show()
```
