```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.special import *
```

```python
x = np.linspace(-4, 4, 1000)

plt.figure(figsize=(8,4))

plt.plot(x, 2*np.abs(jv(1,np.pi*x)/np.pi/x)) # round
plt.plot(x, np.abs(np.sin(np.pi*x)/np.pi/x)) # rectangular

plt.xlim(-4, 4)
plt.ylim(0, 1.2)
plt.xlabel('$\Phi/\Phi_0$')
plt.ylabel('$I_s$ / $I_c$')

plt.legend(['$|2J_1(\pi x)/\pi x|$', '$|\sin(\pi x)/\pi x|$'])

# plt.savefig("Josephson.eps", dpi = 300, format = 'eps')

plt.show()
```

```python
def phi(tau, i):
    beta = np.sqrt(i**2 - 1)
    return 2*np.arctan((beta / i)*np.tan(beta * tau / 2) + 1 / i)

def I(tau, i):
    return np.sin(phi(tau, i))

def V(tau, i):
    return i - I(tau, i)
```

```python
tau = np.linspace(0, 30, 5000)

plt.figure(figsize=(8,4))

plt.plot(tau, phi(tau, 1.08))
plt.plot(tau, phi(tau, 4.5))

plt.xlim(0, 30)
#plt.ylim(0, 1.2)
plt.xlabel('$\\tau$')
plt.ylabel('$\phi(\\tau)$')

plt.legend(['$i = 1.08$', '$i = 4.5$'])

# plt.savefig("phi.eps", dpi = 300, format = 'eps')

plt.show()
```

```python
tau = np.linspace(0, 30, 5000)

plt.figure(figsize=(8,4))

plt.plot(tau, I(tau, 1.08))
plt.plot(tau, I(tau, 4.5))

plt.xlim(0, 30)
#plt.ylim(0, 1.2)
plt.xlabel('$\\tau$')
plt.ylabel('$I_s(\\tau)$ / $I_c$')

plt.legend(['$i = 1.08$', '$i = 4.5$'])

# plt.savefig("I.eps", dpi = 300, format = 'eps')

plt.show()
```

```python
tau = np.linspace(0, 30, 5000)

plt.figure(figsize=(8,4))

plt.plot(tau, V(tau, 1.08))
plt.plot(tau, V(tau, 4.5))

plt.xlim(0, 30)
#plt.ylim(0, 1.2)
plt.xlabel('$\\tau$')
plt.ylabel('$V(\\tau)$ / $I_cR$')

plt.legend(['$i = 1.08$', '$i = 4.5$'])

# plt.savefig("V.eps", dpi = 300, format = 'eps')

plt.show()
```
