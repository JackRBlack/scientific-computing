```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def E(k, u, t):
    (tp, tm) = (t + u, t - u)
    E2 = (tp + tm * np.cos(k))**2 + tm**2 * np.sin(k)**2
    return np.sqrt(E2)
```

```python
u = 0
k = np.linspace(-1, 1, 100)
Ea = E(k * np.pi, u, 1)
Eb = -Ea
plt.figure(figsize=(10,5))
#plt.rcParams.update({'font.size': 20})
plt.plot(k, Ea)
plt.plot(k, Eb)
plt.xlabel("$k / \pi$")
plt.ylabel("energy (arb. unit)")
plt.title("u = " + str(u))
plt.show()
```
