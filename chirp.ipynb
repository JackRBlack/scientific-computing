```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian_chirp(t, t0, width, a):
    sigma = width / 1.665
    return np.exp(-((t - t0) / sigma)**2 * (1 + 1j * a))

def carrier(t, wavelength):
    omega = 2*np.pi * 3e8 / (wavelength * 1e-9) / 1e15
    return np.exp(-1j*omega*t)
```

```python
t = np.linspace(0, 500, 1000)
envelope = gaussian_chirp(t, 250, 100, 0)
envelope_chirped = gaussian_chirp(t, 250, 100, 4)
```

```python
plt.figure(figsize=(14,5))
plt.rcParams.update({'font.size': 20})
plt.plot(t, envelope.real)
plt.plot(t, envelope_chirped.real)
plt.legend(["no chirp (a = 0)", "chirp (a ≠ 0)"])
plt.xlabel("time (fs)")
plt.ylabel("amplitude (arb. unit)")
plt.tight_layout()
#plt.savefig("gaussian_chirp.eps", dpi = 300, format = 'eps')
plt.show()
```

```python
t = np.linspace(0, 500, 500000)
pulse_chirped_p = gaussian_chirp(t, 250, 100, 20) * carrier(t, 1500)
pulse_chirped_n = gaussian_chirp(t, 250, 100,-20) * carrier(t, 1500)

fig, (ax1, ax2) = plt.subplots(2, 1, sharex='col', figsize=(16,9))
ax1.plot(t, pulse_chirped_p.real, color = 'C0')
ax1.legend(["a > 0"])
ax2.plot(t, pulse_chirped_n.real, color = 'C1')
ax2.legend(["a < 0"])
ax2.set_xlabel("time (fs)")
ax1.set_ylabel("amplitude (arb. unit)")
ax2.set_ylabel("amplitude (arb. unit)")
plt.tight_layout()
plt.savefig("pulse_chirp.eps", dpi = 300, format = 'eps')
plt.show()
```
