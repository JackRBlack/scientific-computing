```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 100, 10000)
```

```python
signal = x - x
for i in range(10000):
    if i%50 == 0:
        signal[i] = 1
```

```python
# Plot the figure
plt.figure(figsize=(20,5))
plt.rcParams.update({'font.size': 20})
plt.plot(x, signal)
plt.xlabel("x")
plt.ylabel("signal")
plt.show()
```

```python
n = len(signal)
rSIGNAL = np.fft.rfft(signal, n)
xstep = x[1]
rfreq = np.fft.rfftfreq(n, d=xstep)
```

```python
# Plot the figure
plt.figure(figsize=(20,5))
plt.rcParams.update({'font.size': 20})
plt.plot(rfreq, np.absolute(rSIGNAL))
plt.xlabel("frequency")
plt.ylabel("amplitude of SIGNAL")
plt.show()
```
