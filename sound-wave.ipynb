This project is a demonstration of how harmonic waves superpose. Different superposing coefficients correspond to different tone color.

## 1. Codes

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def wave_generator(freq, time):
    '''
        Generate sine-shape wave with specific frequency.
        
        args:
            freq : [double] frequency in Hz.
            time : [np.array] a time series, e.g. np.linspace(0, 10, 1000).
        
        return:
            wave : [np.array] an array of wave amplitude value.
    '''
    wave = np.sin(2 * np.pi * freq * time)
    return wave

def harmonic_waves_generator(freq_0, harmonic_series, time):
    '''
        Generate a set of harmonic waves with specific fundamental frequency.
        
        args:
            freq_0 : [double] fundamental frequency in Hz.
            harmonic_series : [np.array] an array of amplitudes for different harmonic waves.
            time : [np.array] a time series, e.g. np.linspace(0, 10, 1000).
        
        return:
            waves : [np.array] an array of harmonic waves.
    '''
    waves = []
    n = 1
    for amplitude in harmonic_series:
        wave = amplitude * wave_generator(freq_0 * n, time)
        waves.append(wave)
        n = n + 1
    return np.array(waves)

def plot_harmonic_waves(time, waves):
    '''
        Generate a figure of harmonic waves, along with their superposition.
        
        args:
            time : [np.array] a time series, e.g. np.linspace(0, 10, 1000).
            waves : [np.array] an array of harmonic waves.
            
        return:
            a figure.
    '''
    LEGENDS = []
    n = 1
    wave_T = time * 0
    plt.figure(figsize=(10,5))
    for wave in waves:
        plt.plot(time, wave, '--')
        LEGENDS.append("n = " + str(n))
        wave_T = wave_T + wave
        n = n + 1
    plt.plot(time, wave_T, 'k', linewidth = 2)
    LEGENDS.append("total")
    plt.legend(LEGENDS)
    plt.xlabel("time / s")
    plt.ylabel("amplitude")
    plt.show()
    return
```

## 2. Examples

```python
# Define the time series.
# begin: 0s
# end : 0.002s
# step : 1000
time = np.linspace(0, 0.002, 1000)

# Amplitudes for different harmonics. The summation should be equal to 1.
# For example the following assignment means: 
#     total wave =  60% of fundamental frequency wave + 30% of 2nd harmonic wave + 10% of 3rd harmonic wave.
harmonic_series = np.array([0.6, 0.3, 0.1])

# Calculate all harmonic waves for the given fundamental frequency.
# For example the following assignment means:
#     fundamental frequency = 1000 Hz.
waves = harmonic_waves_generator(1000, harmonic_series, time)

# Generate the figure.
plot_harmonic_waves(time, waves)
```

```python
time = np.linspace(0, 0.002, 1000)
harmonic_series = np.array([0.6, 0.05, 0.2, 0.15])
waves = harmonic_waves_generator(1000, harmonic_series, time)
plot_harmonic_waves(time, waves)
```

```python
time = np.linspace(0, 0.002, 1000)
harmonic_series = np.array([0.5, 0.2, 0.15, 0.1, 0.05])
waves = harmonic_waves_generator(1000, harmonic_series, time)
plot_harmonic_waves(time, waves)
```
