```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as pat
```

```python
def f(t):
    return np.tanh(t/np.sqrt(2))
def psi(x, t):
    return np.exp(-np.abs(x)/t)
```

```python
x = np.linspace(0, 5, 500)
plt.figure(figsize=(8,4))
plt.plot(x, f(x), 'k')
plt.xlabel('x / $\\xi(T)$')
plt.ylabel('$\psi(x)$ / $\psi_o$')
#plt.savefig("GL_op.eps", dpi = 300, format = 'eps')
plt.show()
```

```python
x = np.linspace(0, 5, 500)
plt.figure(figsize=(8,4))
plt.plot(x, f(x + 0.6), 'k')
plt.xlabel('x / $\\xi(T)$')
plt.ylabel('$\psi(x)$ / $\psi_o$')
plt.ylim(-0.05, 1.05)
#plt.savefig("GL_op_1.eps", dpi = 300, format = 'eps')
plt.show()
```

```python
x = np.linspace(0, 5, 500)
y = np.linspace(-2, 0, 500)
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0,0,1,1])
a = pat.Rectangle((0, 0), 5, 1, facecolor = '#D0DFD0')
b = pat.Rectangle((-2, 0), 2, 1, facecolor = '#DFDFD0')
ax.add_patch(a)
ax.add_patch(b)
plt.plot(x, f(x + 0.6),'#000F00')
plt.plot(y, 0.4 * psi(y, 0.3), '#8B4513')
ax.text(1.7, 0.5, 'superconductor', fontsize = 16)
ax.text(-1.7, 0.5, 'normal metal', fontsize = 16)
plt.xlabel('x / $\\xi(T)$')
plt.ylabel('$\psi(x)$ / $\psi_o$')
plt.ylim(-0.05, 1.05)
plt.savefig("GL_op_2.eps", dpi = 300, format = 'eps', bbox_inches='tight')
plt.show()
```

```python
x = np.linspace(0.6, 5, 500)
plt.figure(figsize=(8,4))
plt.plot(x, np.log(x)/np.cosh(x)**2, 'b')
#plt.plot(x, np.tanh(x) * np.log(x), 'k')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['$f(x) = ln(x)sech^2(x)$'])
#plt.xlim(-0, 5.1)
#plt.savefig("BCS_2.eps", dpi = 300, format = 'eps')
plt.show()
```

```python
x = np.linspace(0.001, 2, 500)
plt.figure(figsize=(8,4))
plt.plot(x, np.tanh(x) * np.log(x), 'r')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['$f(x) = tanh(x)ln(x)$'])
#plt.xlim(-0, 5.1)
#plt.savefig("BCS_1.eps", dpi = 300, format = 'eps')
plt.show()
```
