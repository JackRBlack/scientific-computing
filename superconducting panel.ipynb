```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def B_z(x, lambda_L):
    return np.cosh(x/lambda_L) / np.cosh(1/lambda_L)

def J_y(x, lambda_L):
    return - np.sinh(x/lambda_L) / np.cosh(1/lambda_L)
```

```python
x = np.linspace(-1, 1, 500)
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, B_z(x, 0.2))
plt.plot(x, B_z(x, 0.1))
plt.legend(['$L = 5\lambda_L$', '$L = 10\lambda_L$'])
plt.xlabel('x / L')
plt.ylabel('$B_z$ / $B_0$')
#plt.savefig("magnetic_field.eps", dpi = 300, format = 'eps')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, J_y(x, 0.2))
plt.plot(x, J_y(x, 0.1))
plt.legend(['$L = 5\lambda_L$', '$L = 10\lambda_L$'])
plt.xlabel('x / L')
plt.ylabel('$J_y$ / $({B_0}/{\mu_0\lambda_L})$')
#plt.savefig("current_density.eps", dpi = 300, format = 'eps')
plt.show()
```
